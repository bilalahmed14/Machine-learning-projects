```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv("./dataset/tmdb_5000_movies.csv")
credits = pd.read_csv("./dataset/tmdb_5000_credits.csv")
movies.head(2)
```

```python
credits.head(2)
```

```python
print(movies.shape, credits.shape)
```

```python
movies = movies.merge(credits, on='title')
movies.head(2)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(2)
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies1 = movies.copy()
movies1.dropna(inplace=True)
movies1['genres'] = movies.genres.apply(convert)
movies1['keywords'] = movies['keywords'].apply(convert)
movies1['cast'] = movies['cast'].apply(convert)

movies1.head(2)
```

```python
movies1['cast'] = movies1['cast'].apply(lambda x:x[0:3]) # to keep only 3 cast name
movies1.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies1['crew'] = movies.crew.apply(fetch_director)
movies1.head(2)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies1['cast'] = movies1['cast'].apply(collapse)
movies1['crew'] = movies1['crew'].apply(collapse)
movies1['genres'] = movies1['genres'].apply(collapse)
movies1['keywords'] = movies1['keywords'].apply(collapse)
movies1.head(2)
```

```python
movies1['overview'] = movies1['overview'].apply(lambda x:x.split())
movies1['tags'] = movies1['overview'] + movies1['genres'] + movies1['keywords'] + movies1['cast'] + movies1['crew']

movies1.head(2)
```

```python
new = movies1.drop(columns=['overview','genres','keywords','cast','crew'])
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

vector = cv.fit_transform(new['tags']).toarray()
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)    
```

```python
recommend('Gandhi')
```
